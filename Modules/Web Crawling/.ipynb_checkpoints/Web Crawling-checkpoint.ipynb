# Requests Module Python

```python
import requests
```

```python
url = 'http://www.mcut.edu.tw'
htmlfile = requests.get(url)
print(type(htmlfile))
```

```python
if htmlfile.status_code == requests.codes.ok:
    print('Success')
    print(htmlfile.text)
else: 
    print('Failed')
print(len(htmlfile.text))
```

```python
import re

if htmlfile.status_code == requests.codes.ok: 
    pattern = input("Please input a string: ")
    if pattern in htmlfile.text: 
        print('Success')
    else: 
        print('Failed')
    
    name = re.findall(pattern, htmlfile.text)
    
    if name: 
        print('%s showed up %d times' % (pattern, len(name)))
    else: 
        print('%s showed up 0 times' % pattern)
else: 
    print('Download failed')
```

```python
try: 
    htmlfile.raise_for_status()
    print('Downloaded successfully')
except Exception as err: 
    print('Download failed because of %s' % err)
```

```python
url = 'http://www.tenlong.com.tw'
try: 
    htmlfile = requests.get(url)
    print('Success')
except Exception as err: 
    print(err)
    
fn = 'out26_11.txt'
with open(fn, 'wb') as file: 
    for diskStorage in htmlfile.iter_content(10240): 
        size = file.write(diskStorage)
        print(size)
    print(fn)
```

# BeautifulSoup Module Python

```python
import bs4
```

```python
htmlFile = requests.get('http://www.tenlong.com.tw')
objSoup = bs4.BeautifulSoup(htmlFile.text, 'lxml')
print(type(objSoup))
```

```python
r = requests.get("http://www.pyclass.com/example.html", headers={'User-agent': 'Mozilla/5.0 (X11; Ubuntu; Linux x86_64; rv:61.0) Gecko/20100101 Firefox/61.0'})
c = r.content

soup = bs4.BeautifulSoup(c, "html.parser")
all = soup.find('div', {'class':'cities'})

for item in all('p'):
    print(item.text)
```
