## Basic Graphing

```python
import matplotlib.pyplot as plt
```

```python
squares = [1, 4, 9, 16, 25, 36, 49, 64]
seq = [1, 2, 3, 4, 5, 6, 7, 8]
plt.plot(seq, squares, linewidth=3)
#linewidth: optional parameter, change width of line
#plt.axis([0, 8, 0, 70])
#[x1, x2, y1, y2]
plt.title('Test Chart', fontsize=24)
plt.xlabel('Value', fontsize=16)
plt.ylabel('Square', fontsize=16)
#fontsize: optional parameter, change font size of words
plt.tick_params(axis='both', labelsize=12, color='red')
plt.show()
```

```python
data1 = [1, 4, 9, 16, 25, 36, 49, 64]
data2 = [1, 3, 6, 10, 15, 21, 28, 36]
seq = [1, 2, 3, 4, 5, 6, 7, 8]
plt.plot(seq, data1, seq, data2)
plt.title('Test Chart', fontsize=24)
plt.xlabel('x-Value', fontsize=14)
plt.ylabel('y-Value', fontsize=14)
plt.tick_params(axis='both', labelsize=12, color='red')
plt.show()
```

## Matplotlib with Numpy Module

```python
import numpy as np
```

```python
#arange() / linspace() method: 產生相同等劇的陣列
x1 = np.linspace(0, 10, num=111)
print(type(x1), x1)
x2 = np.arange(0, 11, 1)
print(type(x2), x2)
x3 = np.arange(11)
print(type(x3), x3)
```

```python
xpt = np.linspace(0, 10, 500)
ypt1 = np.sin(xpt)
ypt2 = np.cos(xpt)
plt.scatter(xpt, ypt1, color=(0, 1, 0))
plt.scatter(xpt, ypt2)
plt.show()
```

```python
left = -2 * np.pi
right = 2 * np.pi
x = np.linspace(left, right, 100)

f1 = 2 * np.sin(x)
f2 = np.sin(2*x)
f3 = 0.5 * np.sin(x)

plt.plot(x, f1)
plt.plot(x, f2)
plt.plot(x, f3)
plt.show()
```

```python
xpt = np.linspace(0, 5, 500)
ypt = 1 - 0.5 * np.abs(xpt-2)
lwidths = (1 + xpt) ** 2
plt.scatter(xpt, ypt, s = lwidths, color = (0, 1, 0))
plt.show()
```

```python
left = -np.pi
right = np.pi
x = np.linspace(left, right, 100)
y = np.sin(3 * x)

plt.plot(x, y)
plt.fill_between(x, 0, y, color='green', alpha = 0.1)
plt.show()
```

## Random Number Application

```python
num = 100
while True: 
    x = np.random.random(100)
    y = np.random.random(100)
    t = x
    plt.scatter(x, y, s=100, c=t, cmap='brg')
    plt.show()
    yORn = input('Continue? (y or n): ')
    if yORn == 'n' or yORn == 'N': 
        break
```

```python
import random
def loc(index): 
    x_mov = random.choice([-3, 3])
    xloc = x[index-1] + x_mov
    y_mov = random.choice([-5, -1, 1, 5])
    yloc = y[index-1] + y_mov
    x.append(xloc)
    y.append(yloc)

num = 10000
x = [0]
y = [0]
while True: 
    for i in range(1, num):
        loc(i)
    t = x
    plt.scatter(x, y, s=2, c=t, cmap = 'brg')
    plt.show()
    yORn = input('Continue? y or n: ')
    if yORn == 'n' or yORn == 'N': 
        break
    else: 
        x[0] = x[num-1]
        y[0] = y[num-1]
        del x[1:]
        del y[1:]
```

## Bar Graphs: bar() method

```python
votes = [300, 400, 500]
N = len(votes)
x = np.arange(N)
width = 0.35
plt.bar(x, votes, width)

plt.ylabel('The number of votes')
plt.title('The election results')
plt.xticks(x, ('James', 'Peter', 'Norton'))
plt.yticks(np.arange(0, 450, 30))
plt.show()
```

## Pie Chart: pie() method

labels: 圓餅圖項目所組成的串列

colors: 圓餅圖項目顏色所組成的串列，如果省略則用預設顏色

explode: 可設定是否從圓餅圖分離的串列，0表示不分離，一般可用0.1分離，數值越大分離越遠

autopct: 表示項目的百分比格式，基本語法是"%格式%%"

labeldistance: 項目標題與圓餅圖中心的距離是半徑的多少倍

center: 原中心座標

shadow: True表示圓餅圖有陰影

```python
sorts = ['Travel', 'Entertainment', 'Education', 'Transportation', 'Food']
fee = [8000, 2000, 3000, 5000, 6000]
plt.pie(fee, labels=sorts, explode=(0, 0.1, 0, 0, 0), autopct = '%1.2f%%')
plt.show()
```
